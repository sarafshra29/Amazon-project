```python
# ques-5 Boolean columns (is_prime_member, is_prime_eligible, is_festival_sale) contain mixed values: True/False, Yes/No, 1/0, Y/N, and some missing entries. Convert all boolean columns to consistent True/False format.
```

```python
import pandas as pd

df = pd.read_csv("amazon_all_years.csv")

bool_map = {
    "true": True, "false": False,
    "yes": True, "no": False,
    "y": True, "n": False,
    "1": True, "0": False
}

cols = ["is_prime_member","is_prime_eligible","is_festival_sale"]

for col in cols:
    df[col] = df[col].astype(str).str.lower().str.strip().map(bool_map)

df[cols] = df[cols].fillna(False)
```

```python
# quest-7 The delivery_days column contains negative values, text entries like 'Same Day', '1-2 days', and some unrealistic values like 50 days. Clean this column to contain only valid numeric delivery days.
```

```python
import pandas as pd
import re

df = pd.read_csv("amazon_all_years.csv")
```

```python
df["delivery_days"] = df["delivery_days"].astype(str).str.lower().str.strip()
def clean_delivery(x):
    
    # Same day delivery
    if "same" in x:
        return 0
    
    # Range like 1-2 days
    if "-" in x:
        try:
            nums = re.findall(r'\d+', x)
            return sum(map(int, nums)) / len(nums)
        except:
            return None
    
    # Extract number from text like "3 days"
    num = re.findall(r'\d+', x)
    if num:
        return int(num[0])
    
    return None

df["delivery_days"] = df["delivery_days"].apply(clean_delivery)
df.loc[df["delivery_days"] < 0, "delivery_days"] = None
df.loc[df["delivery_days"] > 30, "delivery_days"] = None
df["delivery_days"] = pd.to_numeric(df["delivery_days"])
df["delivery_days"] = df["delivery_days"].fillna(df["delivery_days"].median())
```

```python
df = df.to_csv('amazon_allyearclean.csv')
```

```python
# ques-9 The dataset contains outlier prices where some products show prices 100x higher than expected due to data entry errors (decimal point issues). Identify and correct these outliers using statistical methods and domain knowledge
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("amazon_all_years.csv")

df["original_price_inr"] = pd.to_numeric(df["original_price_inr"], errors="coerce")
```

```python
# handling price outliers (IQR method)
# original price is already in numeric
df['original_price_inr'] = pd.to_numeric(df['original_price_inr'], errors='coerce')
Q1 = df['original_price_inr'].quantile(0.25)
Q3 = df['original_price_inr'].quantile(0.75)

IQR = Q3 - Q1

upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR

outliers = df[(df['original_price_inr'] > upper_limit) | 
              (df['original_price_inr'] < lower_limit)]

median_price = df['original_price_inr'].median()

# any values which are more than 10 times the median is pobably an error while entering data
# so we divide such values by 100
df.loc[df['original_price_inr'] >= median_price * 10,
       'original_price_inr'] = df['original_price_inr'] / 100
```

```python
import pandas as pd
```

```python
df['payment_method'] = df['payment_method'].str.lower().str.strip()

payment_mapping = {

'upi': 'UPI',
'phonepe': 'UPI',
'googlepay': 'UPI',
'gpay': 'UPI',

'credit card': 'Credit Card',
'credit_card': 'Credit Card',
'cc': 'Credit Card',

'cash on delivery': 'Cash on Delivery',
'cod': 'Cash on Delivery',
'c.o.d': 'Cash on Delivery'

}
df['payment_method_clean'] = df['payment_method'].replace(payment_mapping)
print(df['payment_method_clean'].value_counts())

hierarchy = {
'UPI': 'Digital Payment',
'Credit Card': 'Card Payment',
'Cash on Delivery': 'Offline Payment'
}

df['payment_category'] = df['payment_method_clean'].map(hierarchy)
```
